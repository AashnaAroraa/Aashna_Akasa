```python
import sqlite3
import pandas as pd
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta
import pytz
```

```python

conn = sqlite3.connect("akasa_db.sqlite")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
  customer_id TEXT PRIMARY KEY,
  customer_name TEXT,
  mobile_number TEXT UNIQUE,
  region TEXT
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
  order_id TEXT,
  mobile_number TEXT,
  order_date_time TEXT,
  sku_id TEXT,
  sku_count INTEGER,
  total_amount REAL
);
""")

print("Tables created successfully!")
```

```python
tz = pytz.timezone('Asia/Kolkata')

# customers
customers = pd.read_csv("/content/task_DE_new_customers.csv").fillna("")
for _, r in customers.iterrows():
    cursor.execute("""
        INSERT OR REPLACE INTO customers (customer_id, customer_name, mobile_number, region)
        VALUES (?,?,?,?)
    """, (r['customer_id'], r['customer_name'], r['mobile_number'], r['region']))
conn.commit()
print("Customers inserted")

#Orders
tree = ET.parse("/content/task_DE_new_orders.xml")
root = tree.getroot()
for o in root.findall('order'):
    try:
        dt = tz.localize(datetime.fromisoformat(o.findtext('order_date_time','').strip())).astimezone(pytz.utc)
    except Exception:
        continue
    cursor.execute("""
        INSERT INTO orders (order_id, mobile_number, order_date_time, sku_id, sku_count, total_amount)
        VALUES (?,?,?,?,?,?)
    """, (
        o.findtext('order_id','').strip(),
        o.findtext('mobile_number','').strip(),
        dt.strftime('%Y-%m-%d %H:%M:%S'),
        o.findtext('sku_id','').strip(),
        int(o.findtext('sku_count','0')),
        float(o.findtext('total_amount','0'))
    ))
conn.commit()
print("Orders inserted")
```

```python
query = """
SELECT c.customer_id, c.customer_name, COUNT(DISTINCT o.order_id) AS order_count
FROM customers c
JOIN orders o ON c.mobile_number = o.mobile_number
GROUP BY c.customer_id, c.customer_name
HAVING order_count > 1
ORDER BY order_count DESC;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT substr(order_date_time, 1, 7) AS month, COUNT(DISTINCT order_id) AS orders_count
FROM orders
GROUP BY month
ORDER BY month;
"""
pd.read_sql_query(query, conn)
```

```python
query = """
SELECT c.region, SUM(o.total_amount) AS total_revenue
FROM customers c
JOIN orders o ON c.mobile_number = o.mobile_number
GROUP BY c.region
ORDER BY total_revenue DESC;
"""
pd.read_sql_query(query, conn)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta
import pytz
```

```python
customers = pd.read_csv("/content/task_DE_new_customers.csv")
customers = customers.drop_duplicates(subset='mobile_number')
customers = customers.fillna({'region': 'Unknown'})
customers.columns = [c.strip().lower() for c in customers.columns]

print(" Customers cleaned successfully!")
display(customers.head())
```

```python
tree = ET.parse("/content/task_DE_new_orders.xml")
root = tree.getroot()

orders_data = []
for order in root.findall("order"):
    try:
        order_id = order.findtext("order_id", "").strip()
        mobile_number = order.findtext("mobile_number", "").strip()
        sku_id = order.findtext("sku_id", "").strip()
        sku_count = int(order.findtext("sku_count", "0"))
        total_amount = float(order.findtext("total_amount", "0"))

        raw_date = order.findtext("order_date_time", "").strip()
        if raw_date:
            date_obj = datetime.fromisoformat(raw_date)
        else:
            continue
        local_tz = pytz.timezone("Asia/Kolkata")
        utc_date = local_tz.localize(date_obj).astimezone(pytz.utc)
        clean_date = utc_date.strftime("%Y-%m-%d %H:%M:%S")

        orders_data.append({
            "order_id": order_id,
            "mobile_number": mobile_number,
            "order_date_time": clean_date,
            "sku_id": sku_id,
            "sku_count": sku_count,
            "total_amount": total_amount
        })
    except Exception as e:
        print("Skipping bad record:", e)

orders = pd.DataFrame(orders_data)
print(" Orders cleaned successfully!")
display(orders.head())
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
from datetime import datetime
import pytz

customers = pd.read_csv("/content/task_DE_new_customers.csv").fillna("")
customers.columns = [c.strip().lower() for c in customers.columns]
print(" Customers loaded and cleaned!")
display(customers.head())

tree = ET.parse("/content/task_DE_new_orders.xml")
root = tree.getroot()

rows = []
for order in root.findall('order'):
    try:
        dt = datetime.fromisoformat(order.findtext('order_date_time','').strip())
    except Exception:
        continue
    rows.append({
        "order_id": order.findtext('order_id','').strip(),
        "mobile_number": order.findtext('mobile_number','').strip(),
        "order_date_time": dt.strftime('%Y-%m-%d %H:%M:%S'),
        "sku_id": order.findtext('sku_id','').strip(),
        "sku_count": int(order.findtext('sku_count','0')),
        "total_amount": float(order.findtext('total_amount','0'))
    })
orders = pd.DataFrame(rows)
print(" Orders loaded and cleaned!")
display(orders.head())
```
